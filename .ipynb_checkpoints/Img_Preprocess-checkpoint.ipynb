* ë°ì´í„° ì €ì‘ê¶Œ ë° ë°ì´í„° ë¶„ëŸ‰ ë¬¸ì œë¡œ ì¸í•´, íŒ€ì› ì˜ìƒ ì´¬ì˜ í›„ frame ë‹¨ìœ„ë¡œ ì¶”ì¶œí•˜ì—¬ ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ íšë“í•˜ì˜€ìŒ.

```python
import cv2
import os
import random
import numpy as np
from PIL import Image
import PIL.ImageOps  
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img  
import random
import re
import shutil
import copy

random.seed(10)
```

```python
# ì„  ìì„¸ (ê±°ë¶ëª©, ì •ìƒëª©)
video_standing_bad = os.listdir('./origin_video/ì„ _ìì„¸/ê±°ë¶ëª©')
video_standing_good = os.listdir('./origin_video/ì„ _ìì„¸/ì •ìƒëª©')

# ì•‰ì€ ìì„¸ (ê±°ë¶ëª©, ì •ìƒëª©)
video_sitting_bad = os.listdir('./origin_video/ì•‰ì€_ìì„¸/ê±°ë¶ëª©')
video_sitting_good = os.listdir('./origin_video/ì•‰ì€_ìì„¸/ì •ìƒëª©')

video_folders = [video_standing_bad, video_standing_good, video_sitting_bad, video_sitting_good]
print(video_folders)
```

```python
total_save_count = 0

for folder_idx, one_folder in enumerate(video_folders) :
  
  # ë ˆì´ë¸” ì •ë¦¬í•˜ê¸°
  if folder_idx == 0 :
    kor_label = 'ì„ _ìì„¸/ê±°ë¶ëª©'
    eng_label = 'StandingBad'
  elif folder_idx == 1 :
    kor_label = 'ì„ _ìì„¸/ì •ìƒëª©'
    eng_label = 'StandingGood'
  elif folder_idx == 2 :
    kor_label = 'ì•‰ì€_ìì„¸/ê±°ë¶ëª©'
    eng_label = 'SittingBad'
  elif folder_idx == 3 :
    kor_label = 'ì•‰ì€_ìì„¸/ì •ìƒëª©'
    eng_label = 'SittingGood'

  print('\n' + kor_label)

  folder_save_num = 1

  for one_file in one_folder :
    origin_file = './origin_video/' + kor_label + '/' + one_file
    savedir = './frame_cut/' + kor_label
    save_file = savedir + '/' + one_file

    # ì €ì¥í•˜ë ¤ëŠ” í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±í•˜ë¼ëŠ” ëœ»
    if not os.path.exists(savedir) :
      print('mkdir=', savedir)
      os.makedirs(savedir)


    # ì´ë¯¸ ì „ì²˜ë¦¬ ì§„í–‰í–ˆë˜ íŒŒì¼ì´ë©´ skip
    if os.path.exists(save_file) :
      continue

    vidcap = cv2.VideoCapture(origin_file)

    success, image = vidcap.read()

    frame_count = 1
    save_count =1
    success = True

    while success :
      success, image = vidcap.read()

      if image is None or save_count > 10 :
        continue
      
      frame_count += 1

      if frame_count % 20 == 0 :
        image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
        cv2.imwrite(f"{savedir}/{eng_label}_{folder_save_num}.jpg", image)
        print("saved image %d.jpg" % save_count)
        save_count += 1
        total_save_count += 1
        folder_save_num += 1

      if cv2.waitKey(10) == 27:   #27 = ESC// exit ì—­í• 
        break

print(f'ì²˜ë¦¬í•œ ì „ì²´ í”„ë ˆì„ ìˆ˜ : {total_save_count}')
```

## ğŸ”¸ ëª¨ë¸ ë°ì´í„° í´ë” ì •ë¦¬

```python
# ì €ì¥í•˜ë ¤ëŠ” í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±í•˜ë¼ëŠ” ëœ»
savedir_list = ['./model_data/ì„ _ìì„¸/train/bad', './model_data/ì„ _ìì„¸/test/bad',
                './model_data/ì„ _ìì„¸/train/good', './model_data/ì„ _ìì„¸/test/good']

for savedir in savedir_list:
  if not os.path.exists(savedir) :
    print('mkdir=', savedir)
    os.makedirs(savedir)


# ì„  ìì„¸ (ê±°ë¶ëª©, ì •ìƒëª©)
frame_standing_bad = [f"./frame_cut/ì„ _ìì„¸/ê±°ë¶ëª©/{i}" for i in os.listdir('./frame_cut/ì„ _ìì„¸/ê±°ë¶ëª©')]
frame_standing_good = [f"./frame_cut/ì„ _ìì„¸/ì •ìƒëª©/{i}" for i in os.listdir('./frame_cut/ì„ _ìì„¸/ì •ìƒëª©')]

train_frame_standing_bad = frame_standing_bad[:75]
test_frame_standing_bad = frame_standing_bad[75:]

train_frame_standing_good = frame_standing_good[:75]
test_frame_standing_good = frame_standing_good[75:]


standing_frame = [train_frame_standing_bad, test_frame_standing_bad, 
                 train_frame_standing_good, test_frame_standing_good]


# frame_cut ë°ì´í„°ë¶€í„° í´ë” ë‚´ìš©ë¶€í„° ì •ë¦¬í•˜ê¸°
for category_idx, category in enumerate(standing_frame):
  for src in category:
    if category_idx == 0:
      dst = savedir_list[0]
      shutil.copy(src,dst)

    elif category_idx == 1:
      dst = savedir_list[1]
      shutil.copy(src,dst)

    elif category_idx == 2:
      dst = savedir_list[2]
      shutil.copy(src,dst)

    elif category_idx == 3:
      dst = savedir_list[3]
      shutil.copy(src,dst)      


for savedir in savedir_list:
  print(len(os.listdir(savedir)))
```

```python
# â­ frame ì´ë¯¸ì§€ ì„ê¸° ('í˜¼í•©_ìì„¸' ë°ì´í„° í´ë”ë¥¼ ë§Œë“¤ê¸° ìœ„í•œ ë‹¨ê³„)
copy_frame_standing_bad = copy.deepcopy(frame_standing_bad)
random.shuffle(copy_frame_standing_bad)
shuffle_frame_standing_bad = copy_frame_standing_bad[:75]

copy_frame_standing_good = copy.deepcopy(frame_standing_good)
random.shuffle(copy_frame_standing_good)
shuffle_frame_standing_good = copy_frame_standing_good[:75]

copy_frame_sitting_bad = copy.deepcopy(frame_sitting_bad)
random.shuffle(copy_frame_sitting_bad)
shuffle_frame_sitting_bad = copy_frame_sitting_bad[:75]

copy_frame_sitting_good = copy.deepcopy(frame_sitting_good)
shuffle_frame_sitting_good = copy_frame_sitting_good[:75]
random.shuffle(copy_frame_sitting_good)


# train, test ë°ì´í„° ë¶„ë¦¬í•˜ê¸°
train_frame_standing_bad = shuffle_frame_standing_bad[:38]
test_frame_standing_bad = shuffle_frame_standing_bad[38:]

train_frame_standing_good = shuffle_frame_standing_good[:38]
test_frame_standing_good = shuffle_frame_standing_good[38:]

train_frame_sitting_bad = shuffle_frame_sitting_bad[:38]
test_frame_sitting_bad = shuffle_frame_sitting_bad[38:]

train_frame_sitting_good = shuffle_frame_sitting_good[:38]
test_frame_sitting_good = shuffle_frame_sitting_good[38:]

shuffle_frame = [train_frame_standing_bad, test_frame_standing_bad, 
                 train_frame_standing_good, test_frame_standing_good, 
                 train_frame_sitting_bad, test_frame_sitting_bad, 
                 train_frame_sitting_good, test_frame_sitting_good]


# ì €ì¥í•˜ë ¤ëŠ” í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±í•˜ë¼ëŠ” ëœ»
savedir_list = ['./model_data/í˜¼í•©_ìì„¸/train/bad', './model_data/í˜¼í•©_ìì„¸/test/bad', 
                './model_data/í˜¼í•©_ìì„¸/train/good', './model_data/í˜¼í•©_ìì„¸/test/good']

for savedir in savedir_list:
  if not os.path.exists(savedir) :
    print('mkdir=', savedir)
    os.makedirs(savedir)


# frame_cut ë°ì´í„°ë¶€í„° í´ë” ë‚´ìš©ë¶€í„° ì •ë¦¬í•˜ê¸°
for category_idx, category in enumerate(shuffle_frame):
  for src in category:
    if category_idx == 0 or category_idx == 4:
      dst = savedir_list[0]
      shutil.copy(src,dst)

    elif category_idx == 1 or category_idx == 5:
      dst = savedir_list[1]
      shutil.copy(src,dst)

    elif category_idx == 2 or category_idx == 6:
      dst = savedir_list[2]
      shutil.copy(src,dst)

    elif category_idx == 3 or category_idx == 7:
      dst = savedir_list[3]
      shutil.copy(src,dst)      


for savedir in savedir_list:
  print(len(os.listdir(savedir)))
```
