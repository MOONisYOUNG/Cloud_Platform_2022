### ✅ '영상 데이터'에서 '이미지 데이터'로 변환 (frame 추출)



* 데이터 저작권 및 데이터 분량 문제로 인해, 팀원 영상 촬영 후 frame 단위로 추출하여 이미지 데이터를 획득하였음.

```python
import cv2
import os
import random
import numpy as np
from PIL import Image
import PIL.ImageOps  
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img  
import random
import re
import shutil
import copy

random.seed(10)
```

```python
# 선 자세 (거북목, 정상목)
video_standing_bad = os.listdir('./origin_video/선_자세/거북목')
video_standing_good = os.listdir('./origin_video/선_자세/정상목')

# 앉은 자세 (거북목, 정상목)
video_sitting_bad = os.listdir('./origin_video/앉은_자세/거북목')
video_sitting_good = os.listdir('./origin_video/앉은_자세/정상목')

video_folders = [video_standing_bad, video_standing_good, video_sitting_bad, video_sitting_good]
print(video_folders)
```

```python
total_save_count = 0

for folder_idx, one_folder in enumerate(video_folders) :
  
  # 레이블 정리하기
  if folder_idx == 0 :
    kor_label = '선_자세/거북목'
    eng_label = 'StandingBad'
  elif folder_idx == 1 :
    kor_label = '선_자세/정상목'
    eng_label = 'StandingGood'
  elif folder_idx == 2 :
    kor_label = '앉은_자세/거북목'
    eng_label = 'SittingBad'
  elif folder_idx == 3 :
    kor_label = '앉은_자세/정상목'
    eng_label = 'SittingGood'

  print('\n' + kor_label)

  folder_save_num = 1

  for one_file in one_folder :
    origin_file = './origin_video/' + kor_label + '/' + one_file
    savedir = './frame_cut/' + kor_label
    save_file = savedir + '/' + one_file

    # 저장하려는 폴더가 없으면 생성하라는 뜻
    if not os.path.exists(savedir) :
      print('mkdir=', savedir)
      os.makedirs(savedir)


    # 이미 전처리 진행했던 파일이면 skip
    if os.path.exists(save_file) :
      continue

    vidcap = cv2.VideoCapture(origin_file)

    success, image = vidcap.read()

    frame_count = 1
    save_count =1
    success = True

    while success :
      success, image = vidcap.read()

      if image is None or save_count > 10 :
        continue
      
      frame_count += 1

      if frame_count % 20 == 0 :
        image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
        cv2.imwrite(f"{savedir}/{eng_label}_{folder_save_num}.jpg", image)
        print("saved image %d.jpg" % save_count)
        save_count += 1
        total_save_count += 1
        folder_save_num += 1

      if cv2.waitKey(10) == 27:   #27 = ESC// exit 역할
        break

print(f'처리한 전체 프레임 수 : {total_save_count}')
```

## 🔸 모델 데이터 폴더 정리

```python
# 저장하려는 폴더가 없으면 생성하라는 뜻
savedir_list = ['./model_data/선_자세/train/bad', './model_data/선_자세/test/bad',
                './model_data/선_자세/train/good', './model_data/선_자세/test/good']

for savedir in savedir_list:
  if not os.path.exists(savedir) :
    print('mkdir=', savedir)
    os.makedirs(savedir)


# 선 자세 (거북목, 정상목)
frame_standing_bad = [f"./frame_cut/선_자세/거북목/{i}" for i in os.listdir('./frame_cut/선_자세/거북목')]
frame_standing_good = [f"./frame_cut/선_자세/정상목/{i}" for i in os.listdir('./frame_cut/선_자세/정상목')]

train_frame_standing_bad = frame_standing_bad[:75]
test_frame_standing_bad = frame_standing_bad[75:]

train_frame_standing_good = frame_standing_good[:75]
test_frame_standing_good = frame_standing_good[75:]


standing_frame = [train_frame_standing_bad, test_frame_standing_bad, 
                 train_frame_standing_good, test_frame_standing_good]


# frame_cut 데이터부터 폴더 내용부터 정리하기
for category_idx, category in enumerate(standing_frame):
  for src in category:
    if category_idx == 0:
      dst = savedir_list[0]
      shutil.copy(src,dst)

    elif category_idx == 1:
      dst = savedir_list[1]
      shutil.copy(src,dst)

    elif category_idx == 2:
      dst = savedir_list[2]
      shutil.copy(src,dst)

    elif category_idx == 3:
      dst = savedir_list[3]
      shutil.copy(src,dst)      


for savedir in savedir_list:
  print(len(os.listdir(savedir)))
```

```python
# 저장하려는 폴더가 없으면 생성하라는 뜻
savedir_list = ['./model_data/앉은_자세/train/bad', './model_data/앉은_자세/test/bad',
                './model_data/앉은_자세/train/good', './model_data/앉은_자세/test/good']

for savedir in savedir_list:
  if not os.path.exists(savedir) :
    print('mkdir=', savedir)
    os.makedirs(savedir)


# 앉은 자세 (거북목, 정상목)
frame_sitting_bad = [f"./frame_cut/앉은_자세/거북목/{i}" for i in os.listdir('./frame_cut/앉은_자세/거북목')]
frame_sitting_good = [f"./frame_cut/앉은_자세/정상목/{i}" for i in os.listdir('./frame_cut/앉은_자세/정상목')]

train_frame_sitting_bad = frame_sitting_bad[:75]
test_frame_sitting_bad = frame_sitting_bad[75:]

train_frame_sitting_good = frame_sitting_good[:75]
test_frame_sitting_good = frame_sitting_good[75:]


sitting_frame = [train_frame_sitting_bad, test_frame_sitting_bad, 
                 train_frame_sitting_good, test_frame_sitting_good]


# frame_cut 데이터부터 폴더 내용부터 정리하기
for category_idx, category in enumerate(sitting_frame):
  for src in category:
    if category_idx == 0:
      dst = savedir_list[0]
      shutil.copy(src,dst)

    elif category_idx == 1:
      dst = savedir_list[1]
      shutil.copy(src,dst)

    elif category_idx == 2:
      dst = savedir_list[2]
      shutil.copy(src,dst)

    elif category_idx == 3:
      dst = savedir_list[3]
      shutil.copy(src,dst)      


for savedir in savedir_list:
  print(len(os.listdir(savedir)))
```

```python
# ⭐ frame 이미지 섞기 ('혼합_자세' 데이터 폴더를 만들기 위한 단계)
copy_frame_standing_bad = copy.deepcopy(frame_standing_bad)
random.shuffle(copy_frame_standing_bad)
shuffle_frame_standing_bad = copy_frame_standing_bad[:75]

copy_frame_standing_good = copy.deepcopy(frame_standing_good)
random.shuffle(copy_frame_standing_good)
shuffle_frame_standing_good = copy_frame_standing_good[:75]

copy_frame_sitting_bad = copy.deepcopy(frame_sitting_bad)
random.shuffle(copy_frame_sitting_bad)
shuffle_frame_sitting_bad = copy_frame_sitting_bad[:75]

copy_frame_sitting_good = copy.deepcopy(frame_sitting_good)
shuffle_frame_sitting_good = copy_frame_sitting_good[:75]
random.shuffle(copy_frame_sitting_good)


# train, test 데이터 분리하기
train_frame_standing_bad = shuffle_frame_standing_bad[:38]
test_frame_standing_bad = shuffle_frame_standing_bad[38:]

train_frame_standing_good = shuffle_frame_standing_good[:38]
test_frame_standing_good = shuffle_frame_standing_good[38:]

train_frame_sitting_bad = shuffle_frame_sitting_bad[:38]
test_frame_sitting_bad = shuffle_frame_sitting_bad[38:]

train_frame_sitting_good = shuffle_frame_sitting_good[:38]
test_frame_sitting_good = shuffle_frame_sitting_good[38:]

shuffle_frame = [train_frame_standing_bad, test_frame_standing_bad, 
                 train_frame_standing_good, test_frame_standing_good, 
                 train_frame_sitting_bad, test_frame_sitting_bad, 
                 train_frame_sitting_good, test_frame_sitting_good]


# 저장하려는 폴더가 없으면 생성하라는 뜻
savedir_list = ['./model_data/혼합_자세/train/bad', './model_data/혼합_자세/test/bad', 
                './model_data/혼합_자세/train/good', './model_data/혼합_자세/test/good']

for savedir in savedir_list:
  if not os.path.exists(savedir) :
    print('mkdir=', savedir)
    os.makedirs(savedir)


# frame_cut 데이터부터 폴더 내용부터 정리하기
for category_idx, category in enumerate(shuffle_frame):
  for src in category:
    if category_idx == 0 or category_idx == 4:
      dst = savedir_list[0]
      shutil.copy(src,dst)

    elif category_idx == 1 or category_idx == 5:
      dst = savedir_list[1]
      shutil.copy(src,dst)

    elif category_idx == 2 or category_idx == 6:
      dst = savedir_list[2]
      shutil.copy(src,dst)

    elif category_idx == 3 or category_idx == 7:
      dst = savedir_list[3]
      shutil.copy(src,dst)      


for savedir in savedir_list:
  print(len(os.listdir(savedir)))
```

## 🔹 '혼합 모델 데이터' 파일명 정리

```python
train_dir_list = ['./model_data/선_자세/train/bad', './model_data/선_자세/train/good']

test_dir_list = ['./model_data/선_자세/test/bad', './model_data/선_자세/test/good']


# 1차 : '혼합_자세' 모델 데이터 파일명 변경하기
train_bad_file_num = 1
test_bad_file_num = 1
train_good_file_num = 1
test_good_file_num = 1

for savdir in train_dir_list :
  for f_name in os.listdir(savdir):
    if re.search(r"Bad", f_name) or re.search(r"bad", f_name):
      new_f_name = f'bad_train_{train_bad_file_num}.jpg'
      os.rename(f'{savdir}/{f_name}', f'{savdir}/{new_f_name}')
      train_bad_file_num += 1

    elif re.search(r"Good", f_name) or re.search(r"good", f_name):
      new_f_name = f'good_train_{train_good_file_num}.jpg'
      os.rename(f'{savdir}/{f_name}', f'{savdir}/{new_f_name}')
      train_good_file_num += 1

    
for savdir in test_dir_list :
  for f_name in os.listdir(savdir):
    if re.search(r"Bad", f_name) or re.search(r"bad", f_name):
      new_f_name = f'bad_test_{test_bad_file_num}.jpg'
      os.rename(f'{savdir}/{f_name}', f'{savdir}/{new_f_name}')
      test_bad_file_num += 1

    elif re.search(r"Good", f_name) or re.search(r"good", f_name):
      new_f_name = f'good_test_{test_good_file_num}.jpg'
      os.rename(f'{savdir}/{f_name}', f'{savdir}/{new_f_name}')
      test_good_file_num += 1


for savedir in train_dir_list:
  print(len(os.listdir(savedir)))

for savedir in test_dir_list:
  print(len(os.listdir(savedir)))
```
